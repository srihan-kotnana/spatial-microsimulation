```python
def subtractDataframes(one, two):
  cons_error = pd.DataFrame(columns= ['a.0.9', 'a.10.19', 'a.20.29', 'a.30.39', 'a.40.49', 'a.50.59', 'a.60.69', 'a.70.79', 'a.80.89', 'f', 'm', 'armed.forces', 'civilian.employed', 'not.in.labor.force', 'unemployed', 'abroad.house', 'diff.us.house', 'same.house', 'i.0.9999', 'i.10000.14999', 'i.15000.24999', 'i.25000.34999', 'i.35000.49999', 'i.50000.64999', 'i.65000.74999', 'i.75000.99999', 'abroad.born', 'naturalization', 'not.citizen', 'puerto.rico.born', 'us.born', 'divorced', 'married', 'never', 'separated', 'widowed', 'no.schooling', 'private', 'public', 'with.disability', 'without.disability', 'federal', 'local', 'not.working', 'private.no.profit', 'private.profit', 'self.employed.corp', 'self.employed.no.corp', 'state', 'with.insurance', 'without.insurance', 'asian', 'black', 'other', 'two', 'white'])
  for i in range(255):
    row1 = one.iloc[i]
    row2 = two.iloc[i]
    row3 = [abs(a_i - b_i) for a_i, b_i in zip(row1, row2)]
    cons_error.loc[len(cons_error)] = row3
  return cons_error
```

```python
def getMeanList(one):
  avr = [0]*56
  cons_error = pd.DataFrame(columns= ['a.0.9', 'a.10.19', 'a.20.29', 'a.30.39', 'a.40.49', 'a.50.59', 'a.60.69', 'a.70.79', 'a.80.89', 'f', 'm', 'armed.forces', 'civilian.employed', 'not.in.labor.force', 'unemployed', 'abroad.house', 'diff.us.house', 'same.house', 'i.0.9999', 'i.10000.14999', 'i.15000.24999', 'i.25000.34999', 'i.35000.49999', 'i.50000.64999', 'i.65000.74999', 'i.75000.99999', 'abroad.born', 'naturalization', 'not.citizen', 'puerto.rico.born', 'us.born', 'divorced', 'married', 'never', 'separated', 'widowed', 'no.schooling', 'private', 'public', 'with.disability', 'without.disability', 'federal', 'local', 'not.working', 'private.no.profit', 'private.profit', 'self.employed.corp', 'self.employed.no.corp', 'state', 'with.insurance', 'without.insurance', 'asian', 'black', 'other', 'two', 'white'])
  for i in range(255):
    row1 = one.iloc[i]
    for j in range(56):
      avr[j]+=row1[j]
  return [x/255 for x in avr]
```

```python
cons_error_r = subtractDataframes(cons, cons_r)
cons_error_s = subtractDataframes(cons, cons_s)
cons_error_rs = subtractDataframes(cons, cons_rs)
```

```python
rml = getMeanList(cons_error_r)
sml = getMeanList(cons_error_s)
rsml = getMeanList(cons_error_rs)
```

```python
def getGraph(name, initial):
  categories = cons.columns.array
  #print(categories)
  lengths = {
    "Age": 9,
    "Sex": 2,
    "Employment": 4,
    "Housing": 3,
    "Income" : 8,
    "Citizenship": 5,
    "Marital Status": 5,
    "Schooling": 3,
    "Disability" : 2,
    "Work": 8,
    "Insurance": 2,
    "Race": 5
  }
  firsts = {
    "Age": "a.0.9",
    "Sex": "f",
    "Employment": "armed.forces",
    "Housing": "abroad.house",
    "Income" : "i.0.9999",
    "Citizenship": "abroad.born",
    "Marital Status": "divorced",
    "Schooling": "no.schooling",
    "Disability" : "with.disability",
    "Work": "federal",
    "Insurance": "with.insurance",
    "Race": "asian"
  }
  
  plots = []
  caty = pd.Index(categories).get_loc(firsts[name])
  plt.figure(figsize=(7,7))
  for i in range(caty, caty+lengths[name]):
      countError = 0
      maxT = 0
      #print(categories[i])
      cons_cat = np.array(cons[categories[i]])
      if initial == 'Real':
        synth_cat = np.array(cons_r[categories[i]])
      elif initial == 'Synthetic':
        synth_cat = np.array(cons_s[categories[i]])
      else:
        synth_cat = np.array(cons_rs[categories[i]])
      for j in range(len(cons_cat)):
          #print(cons_cat[j],synth_cat[j],cons_error_cat[j])
          plt.plot(cons_cat[j],synth_cat[j],'go',alpha=0.5) 
          
  plt.title(name,size=25)
  plt.xlabel("Real Census Tract Population", size = 20)
  plt.ylabel("Synthetic Census Tract Population \nBased on " + initial + "Data", size = 20)
  fig1 = plt.gcf()
  fig1.savefig('../{fname}_{type}.png'.format(fname=name, type=initial))
  plt.show()
```

```python
getGraph("Age", "Real")
getGraph("Age", "Synthetic")
getGraph("Age", "Real and Synthetic")
```

```python
def doPlot(i1, i2, data, name):
  plt.title(name,size=20)
  plt.bar(colList[i1:i2], data[i1:i2])
  ax1 = plt.subplot()
  ax1.set_xticks(colList[i1:i2])
  ax1.set_xticklabels(colList[i1:i2], rotation=90)
  plt.show()
```
