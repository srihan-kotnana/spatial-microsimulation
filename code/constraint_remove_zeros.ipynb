```python
def replaceZeros(df):
  names = df.columns.values.tolist()
  ndf = pd.DataFrame(columns =df.columns.values.tolist())
  dkeys = dict(zip(names, [i for i in range(len(names))]))
  print(dkeys)
  for index, rows, in df.iterrows():
    nrows = [int(str(item).replace(",", "")) for item in rows]
    maxElement = max(nrows)
    maxName = ""
    for name in names:
      if nrows[dkeys[name]] == maxElement:
        maxName = name
    for name in names:
      if nrows[dkeys[name]] == 0:
        nrows[dkeys[name]]+=1
        nrows[dkeys[maxName]]-=1
    ndf.loc[len(ndf)] = nrows
  return ndf

def printMin(df):
  names = df.columns.values.tolist()
  for index, rows, in df.iterrows():
    nrows = [int(str(item).replace(",", "")) for item in rows]
    print(min(nrows))

def mess(df):
  names = df.columns.values.tolist()
  for index, rows, in df.iterrows():
    rows = [int(str(item).replace(",", "")) for item in rows]
    print(rows)
```

```python
age = replaceZeros(age)
cit = replaceZeros(cit)
dis = replaceZeros(dis)
emp = replaceZeros(emp)
inc = replaceZeros(inc)
ins = replaceZeros(ins)
mar = replaceZeros(mar)
mob = replaceZeros(mob)
rac = replaceZeros(rac)
sch = replaceZeros(sch)
sex = replaceZeros(sex)
wrk = replaceZeros(wrk)
```
