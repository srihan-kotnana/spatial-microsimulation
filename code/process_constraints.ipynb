```python
from google.colab import files
import pandas as pd
import io
import sys
```

```python
uploaded = files.upload()
```

```python
age = pd.read_csv(io.BytesIO(uploaded['age.csv']))
citizenship = pd.read_csv(io.BytesIO(uploaded['citizenship.csv']))
disability = pd.read_csv(io.BytesIO(uploaded['disability.csv']))
employment = pd.read_csv(io.BytesIO(uploaded['employment.csv']))
income = pd.read_csv(io.BytesIO(uploaded['income.csv']))
insurance = pd.read_csv(io.BytesIO(uploaded['insurance.csv']))
marital_status = pd.read_csv(io.BytesIO(uploaded['marital_status.csv']))
mobility = pd.read_csv(io.BytesIO(uploaded['mobility.csv']))
race = pd.read_csv(io.BytesIO(uploaded['race.csv']))
schooling = pd.read_csv(io.BytesIO(uploaded['schooling.csv']))
sex = pd.read_csv(io.BytesIO(uploaded['sex.csv']))
worker = pd.read_csv(io.BytesIO(uploaded['worker.csv']))
```

```python
age.rename(columns = {'id':'geo_id'}, inplace = True)
age.rename(columns = {'Geographic Area Name':'tract_name'}, inplace = True)
```

```python
pop = age["population"]
disability['without.disability'] = pop - disability["with.disability"]
employment['not.in.labor.force'] = pop - (employment['armed.forces'] + employment['unemployed'] + employment['civilian.employed'])
income["0_9999"] = pop- (income["10000_14999"]+income["15000_24999"]+income["25000_34999"]+income["35000_49999"]+income["50000_64999"]+income["65000_74999"]+income["75000_99999"]+income["1000000_"])
income = income[["0_9999", "10000_14999", "15000_24999", "25000_34999", "35000_49999", "50000_64999", "65000_74999", "75000_99999", "1000000_"]]
insurance["without_insurance"] = pop - insurance["with.insurance"]
marital_status["never"] = pop - (marital_status["married"] + marital_status["widowed"] + marital_status["divorced"] + marital_status["separated"])
mobility["same.house"] = mobility["same.house"] + pop - mobility["total.house"]
mobility = mobility[["same.house", "diff.us.house", "abroad.house"]]
schooling["no.schooling"] = pop - schooling["public"] - schooling["private"]
worker["not.working"] = pop - (worker["private.profit"] + worker["private.no.profit"] + worker["local"] + worker["state"] + worker["federal"] + worker["self.employed.no.corp"] + worker["self.employed.corp"])
```
