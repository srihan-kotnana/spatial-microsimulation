```python
from google.colab import files
import pandas as pd
import io
import sys
from termcolor import colored, cprint
```

```python
uploaded = files.upload()
```

```python
inds = pd.read_csv(io.BytesIO(uploaded['real_inds.csv']))
```

```python
inds.isnull().any()
```

```python
min(inds["AGE"])
```

```python
for i in range(99):
  toReplace = ""
  if i >= 90:
    toReplace = "80_89"
  elif i <= 9:
    toReplace = "0_9"
  else:
    firstDig = str(i//10)
    toReplace = firstDig + "0_" + firstDig + "9"
  #print(str(i) + ": " + toReplace
  inds["AGE"] = inds["AGE"].replace(to_replace=i, value = toReplace)
```

```python
inds["SEX"] = inds["SEX"].replace(to_replace=1, value = "m")
inds["SEX"] = inds["SEX"].replace(to_replace=2, value = "f")
```

```python
inds["EMPLOYMENT"] = inds["EMPLOYMENT"].fillna(6.0)
inds["EMPLOYMENT"] = inds["EMPLOYMENT"].replace(to_replace=1, value = "civilian.employed")
inds["EMPLOYMENT"] = inds["EMPLOYMENT"].replace(to_replace=2, value = "civilian.employed")
inds["EMPLOYMENT"] = inds["EMPLOYMENT"].replace(to_replace=3, value = "unemployed")
inds["EMPLOYMENT"] = inds["EMPLOYMENT"].replace(to_replace=4, value = "armed.forces")
inds["EMPLOYMENT"] = inds["EMPLOYMENT"].replace(to_replace=5, value = "armed.forces")
inds["EMPLOYMENT"] = inds["EMPLOYMENT"].replace(to_replace=6, value = "not.in.labor.force")
```

```python
inds["MOBILITY"] = inds["MOBILITY"].fillna(1.0)
inds["MOBILITY"] = inds["MOBILITY"].replace(to_replace=1, value = "same.house")
inds["MOBILITY"] = inds["MOBILITY"].replace(to_replace=2, value = "abroad.house")
inds["MOBILITY"] = inds["MOBILITY"].replace(to_replace=3, value = "diff.us.house")
```

```python
inds["MOBILITY"].unique()
```

```python
def changeIncome(i):
  toReplace = "1000000_"
  if i < 10000:
    toReplace = "0_9999"
  elif i < 15000:
    toReplace = "10000_14999"
  elif i < 25000:
    toReplace = "15000_24999"
  elif i < 35000:
    toReplace = "25000_34999"
  elif i < 50000:
    toReplace = "35000_49999"
  elif i < 65000:
    toReplace = "50000_64999"
  elif i < 75000:
    toReplace = "65000_74999"
  else:
    toReplace = "75000_99999"
  return toReplace
```

```python
inds["INCOME"] = inds["INCOME"].apply(changeIncome)
```

```python
inds["CITIZENSHIP"] = inds["CITIZENSHIP"].replace(to_replace=1, value = "us.born")
inds["CITIZENSHIP"] = inds["CITIZENSHIP"].replace(to_replace=2, value = "puerto.rico.born")
inds["CITIZENSHIP"] = inds["CITIZENSHIP"].replace(to_replace=3, value = "abroad.born")
inds["CITIZENSHIP"] = inds["CITIZENSHIP"].replace(to_replace=4, value = "naturalization")
inds["CITIZENSHIP"] = inds["CITIZENSHIP"].replace(to_replace=5, value = "not.citizen")
```

```python
inds["RACE"].unique()
```

```python
inds["MARITAL_STATUS"] = inds["MARITAL_STATUS"].replace(to_replace=1, value = "married")
inds["MARITAL_STATUS"] = inds["MARITAL_STATUS"].replace(to_replace=2, value = "widowed")
inds["MARITAL_STATUS"] = inds["MARITAL_STATUS"].replace(to_replace=3, value = "divorced")
inds["MARITAL_STATUS"] = inds["MARITAL_STATUS"].replace(to_replace=4, value = "separated")
inds["MARITAL_STATUS"] = inds["MARITAL_STATUS"].replace(to_replace=5, value = "never")
```

```python
inds["SCHOOLING"] = inds["SCHOOLING"].fillna(1)
inds["SCHOOLING"] = inds["SCHOOLING"].replace(to_replace=1, value = "no.schooling")
inds["SCHOOLING"] = inds["SCHOOLING"].replace(to_replace=2, value = "public")
inds["SCHOOLING"] = inds["SCHOOLING"].replace(to_replace=3, value = "private")
```

```python
inds["DISABILITY"] = inds["DISABILITY"].replace(to_replace=1, value = "with.disability")
inds["DISABILITY"] = inds["DISABILITY"].replace(to_replace=2, value = "without.disability")
```

```python
inds["WORK"] = inds["WORK"].fillna(9)
inds["WORK"] = inds["WORK"].replace(to_replace=1, value = "private.profit")
inds["WORK"] = inds["WORK"].replace(to_replace=2, value = "private.no.profit")
inds["WORK"] = inds["WORK"].replace(to_replace=3, value = "local")
inds["WORK"] = inds["WORK"].replace(to_replace=4, value = "state")
inds["WORK"] = inds["WORK"].replace(to_replace=5, value = "federal")
inds["WORK"] = inds["WORK"].replace(to_replace=6, value = "self.employed.no.corp")
inds["WORK"] = inds["WORK"].replace(to_replace=7, value = "self.employed.corp")
inds["WORK"] = inds["WORK"].replace(to_replace=8, value = "not.working")
inds["WORK"] = inds["WORK"].replace(to_replace=9, value = "not.working")
```

```python
inds["INSURANCE"] = inds["INSURANCE"].replace(to_replace=1, value = "with.insurance")
inds["INSURANCE"] = inds["INSURANCE"].replace(to_replace=2, value = "without.insurance")
```

```python
inds["RACE"] = inds["RACE"].replace(to_replace=1, value = "white")
inds["RACE"] = inds["RACE"].replace(to_replace=2, value = "black.aa")
inds["RACE"] = inds["RACE"].replace(to_replace=3, value = "a.ind..nat.ak")
inds["RACE"] = inds["RACE"].replace(to_replace=4, value = "a.ind..nat.ak")
inds["RACE"] = inds["RACE"].replace(to_replace=5, value = "a.ind..nat.ak")
inds["RACE"] = inds["RACE"].replace(to_replace=6, value = "asian")
inds["RACE"] = inds["RACE"].replace(to_replace=7, value = "nat.hi.pt.isl")
inds["RACE"] = inds["RACE"].replace(to_replace=8, value = "other")
inds["RACE"] = inds["RACE"].replace(to_replace=9, value = "two")
```
